```python
# Initial imports
import pandas as pd
from pathlib import Path
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import sqlalchemy as sql
from sqlalchemy import create_engine
```

```python
df = pd.read_sql_table('joined', con=engine)
df
```

```python
#encoding string variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2 = df.copy()
df2['genre'] = le.fit_transform(df2['genre'])
```

```python
df3 = df2.copy()
df3['publisher'] = le.fit_transform(df3['publisher'])
```

```python
df4 = df3.copy()
df4['platform'] = le.fit_transform(df4['platform'])
```

```python
df4.head(10)
```

```python
#Drop null Values
df_final = df4[['global_sales', 'critic_score','publisher','genre','platform']].dropna()
```

```python
df_final
```

```python
#Scale the data
from sklearn.preprocessing import StandardScaler
data_scaler = StandardScaler()
scaled_df = pd.DataFrame(data_scaler.fit_transform(df_final))
```

```python
scaled_df
```

```python
#Define variables 
X = scaled_df[[1,2,3,4]]
y = scaled_df[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Creating Linear Regression instance.
model = LinearRegression()
# Fitting the model.
model = model.fit(X_train_scaled, y_train)
```

```python
model.score(X_test, y_test)
```
